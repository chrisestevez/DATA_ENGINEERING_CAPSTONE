```python
import requests
import json
import pandas as pd
import glob
import os
import datetime

def urljoin(*parts):
    return ''.join([p for p in parts])

def noaa_api(base_url="https://www.ncei.noaa.gov/access/services/data/v1?",
        dataset="daily-summaries",
        data_type="TMAX,TMIN,TAVG,AWND,WT01,PRCP,SNOW,AWND",
        stations="USW00094728", start_date="", end_date="")->pd.DataFrame:
    
    url = urljoin(base_url,"dataset=",dataset, "&dataTypes=",data_type,
    "&stations=",stations,"&startDate=",start_date,"&endDate=",end_date,
    "&includeAttributes=0&units=standard&format=json")

    resp = requests.get(url)

    if resp.status_code == 200:
        json_data = resp.json() 
        return pd.DataFrame(json_data)
    else:
        print("Request Failed")
```

```python
files = glob.glob('data/motor_collision_data/*')
format = '%Y-%m-%d'
observation_dates = []
for file in files:
    base_date = os.path.basename(file)
    date_extracted = str(base_date).replace("_","-")
    date_formatted = datetime.datetime.strptime(date_extracted, format).date()
    observation_dates.append(date_formatted)
min_date = min(observation_dates).strftime(format)
max_date = max(observation_dates).strftime(format)
```

```python
weather_data = noaa_api(start_date=min_date,end_date=max_date)
```
